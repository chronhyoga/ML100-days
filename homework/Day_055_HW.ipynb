### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
estimators = [('k_means_3', KMeans(n_clusters=3)), 
              ('k_means_ninit_1', KMeans(n_clusters=3, n_init=1, init='random')),
             ('k_means_ninit_5', KMeans(n_clusters=3, n_init=5, init='random')), 
             ('k_means_ninit_10', KMeans(n_clusters=3, n_init=10, init='random'))]
```

```python
fignum = 1
titles = ['3 clusters', '3 clusters_ninit_1', 'clusters_ninit_5', 'clusters_ninit_10']
for name, est in estimators:
    est.fit(X)
    labels = est.labels_
    fig = plt.figure(fignum)
    ax = Axes3D(fig)
    ax.scatter(X[:,1], X[:,2], X[:,3], c=labels)
    ax.set_title(titles[fignum - 1])
    fignum = fignum + 1
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,1], X[:,2], X[:,3], c=y)
ax.set_title('Truth')
```
