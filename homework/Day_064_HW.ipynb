```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cv2.cvtColor is an OpenCV function to convert images to different color spaces
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor , minNeighbors):
 #just making a copy of image passed, so that passed image is not changed 
 img_copy = colored_img.copy()          
 
 #convert the test image to gray image as opencv face detector expects gray images
 gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)          
 
 #let's detect multiscale (some images may be closer to camera than others) images
 faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=minNeighbors);          
 
 #go over list of faces and draw them as rectangles on original colored img
 for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 3)              
 
 return len(faces), img_copy
```

```python
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
test5 =cv2.imread('data/test5.jpg')
for i in [1.2 ,1.3]:
    for j in [1,3,5]:
        n_Faces, faces_detected_img = detect_faces(haar_face_cascade, test5, scaleFactor = i, minNeighbors = j)
        print(f"scaleFactor={i}, minNeighbors = {j}, n_Faces = {n_Faces}")
```

```python
n_Faces, faces_detected_img = detect_faces(haar_face_cascade, test5, scaleFactor = 1.2, minNeighbors = 3)
#plt.imshow(convertToRGB(faces_detected_img))
dpi = 80
margin = 0.05 # (5% of the width/height of the figure...)
xpixels, ypixels = 1000, 1000
figsize = (1 + margin) * ypixels / dpi, (1 + margin) * xpixels / dpi
fig = plt.figure(figsize=figsize, dpi=dpi)
ax = fig.add_axes([margin, margin, 1 - 2*margin, 1 - 2*margin])
ax.imshow(convertToRGB(faces_detected_img))
plt.show()
```

```python
lbp_face_cascade  = cv2.CascadeClassifier('data/lbpcascade_frontalface_improved.xml')
test5 =cv2.imread('data/test5.jpg')
for i in [1.1, 1.2 ,1.3, 1.4]:
    for j in [1,3,5,7]:
        n_Faces, faces_detected_img = detect_faces(lbp_face_cascade , test5, scaleFactor = i, minNeighbors = j)
        print(f"scaleFactor={i}, minNeighbors = {j}, n_Faces = {n_Faces}")
```
